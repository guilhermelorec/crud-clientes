```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, GlobalMaxPool1D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD, Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_excel('/content/historicoPetroleo.xlsx')
df['Preço'] = df['Preço'].astype(str).str.replace(',', '.').astype(float)
df
```

```python
# @title Preço

from matplotlib import pyplot as plt
df['Preço'].plot(kind='line', figsize=(12, 6), title='Preço')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.legend()
plt.grid(True)
plt.show()
```

```python
dadosNormalizados = df[['Preço']]
scaler = StandardScaler()
dadosNormalizados = scaler.fit_transform(dadosNormalizados)
```

```python
def criarSequencia (dados, janela = 5):
  x = []
  y = []
  for i in range(janela, len(dados)):
    x.append(dados[i-janela:i])
    y.append(dados[i])
  return np.array(x), np.array(y)
```

```python
janela = 5
x, y = criarSequencia(dadosNormalizados, janela)
```

```python
tamanhoTreino = int(len(x) * 0.8)
xTreino, xTeste = x[:tamanhoTreino], x[tamanhoTreino:]
yTreino, yTeste = y[:tamanhoTreino], y[tamanhoTreino:]
```

```python
xTreino = np.reshape(xTreino, (xTreino.shape[0], xTreino.shape[1], 1))
xTeste = np.reshape(xTeste, (xTeste.shape[0], xTeste.shape[1], 1))
```

```python
modelo = Sequential()
modelo.add(LSTM(64, return_sequences=False, input_shape=(janela, 1)))
modelo.add(Dense(1))
modelo.compile(optimizer='adam', loss='mean_squared_error')
```

```python
historico = modelo.fit(xTreino, yTreino, epochs = 50, batch_size = 5, verbose = 1)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(historico.history['loss'], marker='o')
plt.title('Perda durante o treinamento')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.grid(True)
plt.show()
```

```python
previsoes = modelo.predict(xTeste)
previsoes = scaler.inverse_transform(previsoes)
yTesteReal = scaler.inverse_transform(yTeste)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(yTesteReal, label='Preço Real')
plt.plot(previsoes, label='Preço Previsão')
plt.title(f'Previsão do preço do Petróleo')
plt.xlabel('Tempo')
plt.ylabel('Preço')
plt.legend()
plt.grid(True)
plt.show()
```
